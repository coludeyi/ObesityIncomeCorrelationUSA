```python
import pandas as pd
```

```python
import numpy as np
```

```python
import matplotlib.pyplot as plt
```

```python
file = pd.read_csv("ObesityPrevalenceRate.csv")
```

```python
states = file["State"].tolist()
obesity_rate = file["OverallObesityRate"].tolist()
white_obesity_rate = file["White"].tolist()
black_obesity_rate = file["Black"].tolist()
hispanic_obesity_rate = file["Hispanic"].tolist()
average_income = file["AverageIncome"].tolist()
```

```python
income = []

for i in average_income:
    num = i.replace(',', '')
    income.append(float(num))
```

```python
plt.xlabel('Average Income')
plt.ylabel('Overall Obesity Rate %')
plt.title('Correlation between Average Income & Obesity Rate in USA')
plt.ylim(min(obesity_rate)-1,max(obesity_rate)+1)
plt.scatter(income, obesity_rate)
plt.show()
```

```python
plt.ylabel("Obesity Rate for White ethnic group")
plt.title('Correlation between Average Income & Obesity Rate for White ethnic group in USA')
plt.ylim(min(white_obesity_rate)-1,max(white_obesity_rate)+1)
plt.scatter(income, white_obesity_rate)
plt.show()
```

```python
plt.ylabel("Obesity Rate for Black ethnic group")
plt.title('Correlation between Average Income & Obesity Rate for Black ethnic group in USA')
smallest = -1
for i in sorted(black_obesity_rate):
    if i > -1:
        smallest = i
        break

plt.ylim(smallest,max(black_obesity_rate)+1)
plt.scatter(income, black_obesity_rate)
plt.show()
```

```python
plt.ylabel("Obesity Rate for Hispanic ethnic group")
plt.title('Correlation between Average Income & Obesity Rate for Hispanic ethnic group in USA')
plt.ylim(min(hispanic_obesity_rate)-1,max(hispanic_obesity_rate)+1)
plt.scatter(income, hispanic_obesity_rate)
plt.show()
```
